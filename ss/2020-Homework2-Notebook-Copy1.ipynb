#### 1.2.3.3 Theil-U and Cramers-V plots

With both plots the correlation ranges from 0 to 1, however, the correlation can never be negative.

**Theil-U** or the uncertainty coefficient is an assymetric view of the correlation between two categorical features. The plot can indicate how the correlation between two features going one way (i.e. x -> y) does not neccessarily mean that the correlation is equal in the oposite direction. 

**Cramers-V** on the other hand shows a symmetric correlation plot between the features. for this case the correllation between x and y is the same as y and x. Cramers-V is based on Pearson’s Chi-Square Test.

Source: https://towardsdatascience.com/the-search-for-categorical-correlation-a1cf7f1888c9

#### 1.2.3.4 plot pairwise interaction

Obtain the lists of patients who lived and died

Function to gather the information for each pairwise plot


```python
def categroicalPlot(featureName, targetFeature, df_cat):
    """
        Function to generate two bar plots of a categorical feature and target feature pairwise
        Assumes:
        Target feature is a binary construct
        Target feature has only 1 (positive (p)) or 0 (negative(n))
        FeatureName is the name in the dataframe
        df_cat contains the target feature and the categorical features
        
    """
    # Determine which rows have positive target
    p_target_true = df_cat.apply(lambda x: True if (x[targetFeature] == 1) else False, axis=1)
    p_target_list = p_target_true.tolist()

    # Determine which rows have negative target
    n_target_true = df_cat.apply(lambda x: True if (x[targetFeature] == 0) else False, axis=1)
    n_target_list = n_target_true.tolist()
    
    # Separate out the rows with a positive target for the feature
    p_target_feature = df_cat[featureName].iloc[p_target_list]
    p_target_feature_count = p_target_feature.value_counts(sort=False)

    # Separate out the rows with negative target for the feature
    n_target_feature = df_cat[featureName].iloc[n_target_list]
    n_target_feature_count = n_target_feature.value_counts(sort=False)

    # Add the feature positive target counts to a dictionary
    p_target_counts_dict = p_target_feature_count.to_dict()

    # Add the feature negative target counts to a dictionary
    n_target_counts_dict = n_target_feature_count.to_dict()

    # Initialise feature list
    feature_list = pd.Series(df_cat[featureName])
    feature_list = pd.unique(feature_list)
    feature_list.sort()

    # Initialise the target vs feature dictionary
    target_vs_feature_dict = {}
    
    # for each feature pair the positive target feature with the negaitve target feature
    for feature in feature_list:
        # If the feature doesn't exist in the dictionary create it
        if feature not in p_target_counts_dict: p_target_counts_dict[feature] = 0
        if feature not in n_target_counts_dict: n_target_counts_dict[feature] = 0
        target_vs_feature_dict[feature] = [p_target_counts_dict[feature],n_target_counts_dict[feature]]
    
    # turn the dictionary into a dataframe
    target_vs_feature_df = pd.DataFrame.from_dict(target_vs_feature_dict)

    # assign the columns to the respective x values
    x1 = target_vs_feature_df.iloc[0, 0:]
    x2 = target_vs_feature_df.iloc[1, 0:]

    # assign the value to p_pos
    p_pos = np.arange(len(x1))
    p_pos = [x for x in p_pos]
    
    return x1, x2, p_pos
```

#### Create a function to call the regression metrics later on

```python
#This function is used repeatedly to compute all metrics
def printRegressionMetrics(testActualVal, predictions):
    #Regression evaluation measures
    print('\n==============================================================================')
    print("Mean Absolute Error: ", metrics.mean_absolute_error(testActualVal, predictions))
    print("Mean Squared Error: ", metrics.mean_squared_error(testActualVal, predictions))
    print("Root Mean Squared Error: ", metrics.mean_squared_error(testActualVal, predictions)**0.5)
    print("R2 score: ", metrics.r2_score(testActualVal, predictions))
    print('\n==============================================================================')
```

#### Create a function to call the classification metrics later on

```python
def printClassificationMetrics(testActualVal, predictions):
    #Classification evaluation measures
    print('\n==============================================================================')
    print("Accuracy: ", metrics.accuracy_score(testActualVal, predictions))
    print("Confusion matrix: \n", metrics.confusion_matrix(testActualVal, predictions))
    print("Classification report:\n ", metrics.classification_report(testActualVal, predictions, zero_division=0))
    print('\n==============================================================================')
```

### 2.7 Cross Validation

```python
def cross_val_LinReg(X, y, cv=3, scoring='accuracy'):
    """Functions to carry out cross validation on the linear regression model
    Default number of validations is 3. The randon state will be updated 
    at each iteration to allow our results to be repeated"""
    
    # store results
    results = []
    # evaluate cv times and append to results
    for i in range(cv):
        # set up train test split
        X_train_cv, X_test_cv, y_train_cv, y_test_cv = train_test_split(X, y, random_state=i , test_size=0.3)
        # generate model
        multiple_linear_reg = LinearRegression().fit(X_train_cv, y_train_cv)
        # threshold
        multiple_linear_reg_predictions = (multiple_linear_reg.predict(X_test_cv) >= 0.5) * 1.0
        # calc score
        if scoring=='accuracy':
            score = metrics.accuracy_score(y_test_cv, multiple_linear_reg_predictions)
        elif scoring=='precision':
            score = metrics.precision_score(y_test_cv, multiple_linear_reg_predictions, zero_division = 0)
        elif scoring=='f1':
            score = metrics.f1_score(y_test_cv, multiple_linear_reg_predictions, zero_division=0)
        elif scoring=='recall':
            score = metrics.recall_score(y_test_cv, multiple_linear_reg_predictions, zero_division=0)
        # append to results
        results.append(score)
    return results
```

```python
def cross_val_LinReg_DF(X,y):
    """Function to perform cross validation and store results 
    in dataframe. Cross validation looks at accuracy, precision, 
    recall, f1. Returns a dataframe with results"""

    linRegResults = {}
    # metrics to test against
    test_metrics = ['accuracy','precision','recall', 'f1']
    
    for metric in test_metrics:
        # generate test results
        result = cross_val_LinReg(X, y, cv=10, scoring=metric)
        length = len(result)
        # store result in dict
        linRegResults[metric] = sum(result)/length

    # create dataframe with results
    LinRegDF = pd.DataFrame.from_dict(linRegResults, orient='index', columns=['Linear_Regression'])
    
    return LinRegDF
```

```python
X = pd.get_dummies(X, drop_first=True)
y = pd.get_dummies(y, drop_first=True)
```

```python
linRegDF = cross_val_LinReg_DF(X,y)
print(f"Mean results from 10 fold cross validation are:")
linRegDF
```

The mean results from the 10 fold cross validation are significantly low. From these results it is theorised that the model is not determining whether the patient will die from covid very well. Although the accuracy is 96% the rest of the indicators are zero showing that although the model is good at determining who is likely to survive. It has not learned how to determine if someone is likely to die.
